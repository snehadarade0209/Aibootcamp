```python
# Import necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Ensure necessary NLTK data is downloaded
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Provided text
text = """
The Indian Premier League (IPL) is a popular professional Twenty20 cricket league in India, established in 2008. It features ten teams from various cities across the country, showcasing a mix of Indian and international cricket talent. The fast-paced nature of IPL matches, filled with thrilling last-ball finishes and high-scoring games, has garnered immense fan following. IPL is more than just cricket; it blends sports with entertainment, involving celebrity owners, grand opening ceremonies, and energetic cheerleaders. The league also provides a significant platform for young, emerging players to shine. With substantial investments, sponsorships, and media attention, the IPL has become a global cricketing extravaganza, beloved by fans worldwide.
"""
print(text)
```

```python
# Tokenization of words
words = word_tokenize(text.lower())  # Convert to lowercase for uniformity
print(words)
```

```python
# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
# The general syntax for a list comprehension is: [expression for item in iterable if condition]
print(filtered_words)
```

```python
# Frequency distribution of words
word_freq = Counter(filtered_words)
print(word_freq)
```

```python
# Create a WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
```

```python
# Plotting the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.show()
```
